## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
#Droping Unnecessary columns
#Droping Columns like 'ID', 'Name', and 'SSN' as they are unique for each record but do not provide any meaningful insights or pattern for analyis, Hence, we removed it.
df.drop(['ID', 'Name','SSN'], axis = 1, inplace = True)


#Check for duplicate rows
#DUplicate rows can lead to biased analysis and incorrct insights, We check for dupicate value if any we found, we remove them.
print('Number of duplicate rows:', df.duplicated().sum)
#df.drop_duplicates(inplace = True)

#Checking for missing values
#Even though the dateset looked clean earlier, it is always good practice to recheck as missing value can cause erros in visualization.
print('Total missing values:', df.isnull().sum().sum())
````

````python
#handling missing value using user defined function and replacing it with the median.
#It will give those columns which has numerical values.
num_cols = df.select_dtypes(include = ['float64', 'int64']).columns
#using for loops
for col in num_cols:
  if df[col].isnull().sum()>0:   #using condition if column have missing value and it is greater then 0
    df[col].fillna(df[col].median(), inplace = True)  #using fillna function to fill the missing values with median of that column
    print(f'Filled missing values in{col} with median.') #printing that missing values are sucessfully filled



#It will give those columns which has categorical columns
cat_cols = df.select_dtypes(['object']).columns

for col in cat_cols:     #using condition if column have missing value and it is greater then 0
  if df[col].isnull().sum()>0:#using condition if column have missing value and it is greater then 0
    df[col].fillna(df[col].mode()[0], inplace = True)#using fillna function to fill the missing values with mode of that column

  print(f'Filled missing values in {col} with mode') #printing that missing values are sucessfully filled

#Rechecking to confim no missing values remian
print('Missing value after cleaning', df.isnull().sum().sum())
````

### What all manipulations have you done and insights you found?

During the data cleaning process, the following manipulations were performed to make the dataset analysis-ready:

* Filled missing values in Occupation using the most frequent value (mode).

* Filled missing values in Type_of_Loan using the most frequent value (mode).

* Filled missing values in Credit_Mix using the most frequent value (mode).

* Filled missing values in Payment_of_Min_Amount using the most frequent value (mode).

* Filled missing values in Payment_Behaviour using the most frequent value (mode).

* Filled missing values in Credit_Score using the most frequent value (mode).

After performing these operations:

* All missing values were successfully handled.

* No null or missing entries remain in the dataset (Total missing values = 0).

Insights from this step:

* Most missing values occurred in categorical columns, not numerical ones — showing that the dataset mainly lacked descriptive information, not financial data.

* Since we used the mode, the filled values maintain the original data distribution without introducing outliers.

* The dataset is now completely clean and consistent, ready for Exploratory Data Analysis (EDA).

#### Chart - 11 Bubble Chart (Occupation vs Annual Income vs Monthly Balance)

````python
# Chart - 11 visualization code
grouped = df.groupby('Occupation').agg({
    'Annual_Income': 'mean',
    'Monthly_Balance': 'mean',
    'Occupation' : 'count'
}).rename(columns = {'Occupation': 'Count'}).reset_index()

plt.figure(figsize=(10,6))
plt.scatter(
    grouped['Annual_Income'],
    grouped['Monthly_Balance'],
    s=grouped['Count'] *20,
    alpha = 0.6
)

for i in range(len(grouped)):
  plt.text(
      grouped['Annual_Income'][i],
      grouped['Monthly_Balance'][i],
      grouped['Occupation'][i]
  )

plt.xlabel('Average Annual Income')
plt.ylabel('Average Monthly Balance')
plt.title('Bubble Chart: Occupation vs Income vs Balance')
plt.show()
````
